```python
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth, association_rules

df = pd.read_csv('supermarket.csv')

df.head()  
```

```python
#df.info
```

```python
# Replace '?' with 0 and 't' with 1 for the analysis
df.replace('?', 0, inplace=True)
df.replace('t', 1, inplace=True)
```

```python
df.head(10)
```

```python
# Drop the 'total' column as it does not represent itemsets
df.drop(columns=['total'], inplace=True)
```

```python
# Apply the fpgrowth algorithm
frequent_itemsets = fpgrowth(df, min_support=0.3, use_colnames=True)
display(frequent_itemsets)
```

```python
# Generate association rules
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
```

```python
# Filter rules by confidence > 70%
filtered_rules = rules[rules['confidence'] > 0.7]

# Sort the filtered rules by lift
sorted_filtered_rules = filtered_rules.sort_values(by="lift", ascending=False)
```

```python

top_100_rules = sorted_filtered_rules.head(100)
```

```python

print(top_100_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python

df_top_100_rules = top_100_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```
